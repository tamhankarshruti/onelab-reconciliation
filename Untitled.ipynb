```python
import json

with open("reconciliation_report.json") as f:
    report = json.load(f)

print("=== SUMMARY ===")
for k, v in report["summary"].items():
    print(f"  {k}: {v}")

print("\n=== TOTALS CHECK ===")
for k, v in report["totals_check"].items():
    print(f"  {k}: {v}")

print(f"\n=== ISSUES FOUND: {len(report['issues'])} ===")
for i, issue in enumerate(report["issues"], 1):
    print(f"\n  [{i}] {issue['gap_type']}")
    print(f"      TXN: {issue['transaction_id']}")
    print(f"      Detail: {issue['detail']}")
```
